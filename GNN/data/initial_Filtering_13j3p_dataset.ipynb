```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

# For SIGNAL

```python
signal = pd.read_csv("sig_200k_minpt20_13jets.txt", sep = '\t')
```

```python
signal = signal.fillna(0)
sig_columnsTODrop = list(signal.iloc[:5,2:37].columns)
signal = signal.drop(sig_columnsTODrop, axis = 1)
```

```python
list(signal.columns)
```

```python
pbar = tqdm(desc = 'Progress', total = len(signal))
for i, event in signal.iterrows():
    for jet_num in range(1,14):
        jetName = 'jet' + str(jet_num)
        if ((event[jetName + '_Eta'] < -2.5) | (event[jetName + '_Eta'] > 2.5)):
            signal.loc[i, jetName + '_Eta'] = np.nan
            signal.loc[i, jetName + '_Phi'] = np.nan
            signal.loc[i, jetName + '_pT'] = np.nan
            signal.loc[i, jetName + '_Px'] = np.nan
            signal.loc[i, jetName + '_Py'] = np.nan
            signal.loc[i, jetName + '_Pz'] = np.nan
            signal.loc[i, jetName + '_E'] = np.nan
    pbar.update(1)
```

```python
signal_copy = signal.copy()
for i in range(1,14):
    signal_copy = signal_copy.drop(['jet' + str(i) + '_btag'], axis = 1)
signal_copy = signal_copy.drop(['eventno', 'jetmultiplicity'], axis = 1)


all_zero_or_nan_mask = ((signal_copy == 0) | signal_copy.isna()).all(axis=1)
zero_or_nan_rows = signal[all_zero_or_nan_mask]
```

```python
signal = signal.drop(zero_or_nan_rows.index)
```

```python
signal = signal.drop(signal[signal['jetmultiplicity'] == 0].index)
```
